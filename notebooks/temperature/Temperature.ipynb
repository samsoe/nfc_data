```python
import pandas as pd
```

```python
# /Users/esamsoe/Google Drive/MPG Ranch/Weather Data/Orchard House Archive
src = 'src/orchard_house-export-full_history_001-190422.csv'
```

```python
df = pd.read_csv(src, parse_dates=['Date Time '])
```

```python
df.info()
```

```python
# take space out of Date Time column name
df.rename(columns = {'Date Time ':'datetime'}, inplace=True)
```

```python
df.columns.tolist()
```

```python
# set index to datetime
df = df.set_index('datetime')
```

```python
# build temperature column list
select_temp = [
    'Day of the Week ',
    'Outside Temperature TX1 (°F)',
    'High Outside Temperature TX1 (°F)',
    'Low Outside Temperature TX1 (°F)',
    'Outside Temperature TX2 (°F)',
    'High Outside Temperature TX2 (°F)',
    'Low Outside Temperature TX2 (°F)',
    'Outside Temperature TX3 (°F)',
    'High Outside Temperature TX3 (°F)',
    'Low Outside Temperature TX3 (°F)',
    'Outside Temperature TX4 (°F)',
    'High Outside Temperature TX4 (°F)',
    'Low Outside Temperature TX4 (°F)',
    'Outside Temperature TX5 (°F)',
    'High Outside Temperature TX5 (°F)',
    'Low Outside Temperature TX5 (°F)',
    'Outside Temperature TX6 (°F)',
    'High Outside Temperature TX6 (°F)',
    'Low Outside Temperature TX6 (°F)'
 ]
```

```python
# select for temperature columns
temp = df.loc[:, select_temp].copy()
```

```python
temp.info()
```

```python
temp.describe()
```

```python
temp.columns.tolist()
```

```python
temp.info()
```

```python
# rename columns
temp.columns = [
    'Day of the Week ',
    'Outside Temperature Baldy Summit (°F)',
    'High Outside Temperature Baldy Summit (°F)',
    'Low Outside Temperature Baldy Summit (°F)',
    'Outside Temperature Baldy Draw (°F)',
    'High Outside Temperature Baldy Draw (°F)',
    'Low Outside Temperature Baldy Draw (°F)',
    'Outside Temperature Sanfoin Bench (°F)',
    'High Outside Temperature Sanfoin Bench (°F)',
    'Low Outside Temperature Sanfoin Bench (°F)',
    'Outside Temperature South Baldy (°F)',
    'High Outside Temperature South Baldy (°F)',
    'Low Outside Temperature South Baldy (°F)',
    'Outside Temperature Orchard House (°F)',
    'High Outside Temperature Orchard House (°F)',
    'Low Outside Temperature Orchard House (°F)',
    'Outside Temperature Indian Ridge (°F)',
    'High Outside Temperature Indian Ridge (°F)',
    'Low Outside Temperature Indian Ridge (°F)',
 ]
```

```python
temp.info()
```

```python
output = 'output/orchard_house-temp_hourly.csv'
temp.to_csv(output)
```

```python
temp_daily_mean = temp.groupby(pd.Grouper(freq='D')).mean().round(1)
```

```python
temp_daily_mean = temp_daily_mean[[
    'Outside Temperature Baldy Summit (°F)',
    'Outside Temperature Baldy Draw (°F)',
    'Outside Temperature Sanfoin Bench (°F)',
    'Outside Temperature South Baldy (°F)',
    'Outside Temperature Orchard House (°F)',
    'Outside Temperature Indian Ridge (°F)',
 ]
]
```

```python
output = 'output/orchard_house-temp_daily_mean.csv'
temp_daily_mean.to_csv(output)
```

```python
temp_daily_mean
```

```python
# Monthly Max by Station
temp_monthly_mean = temp.groupby(pd.Grouper(freq='M')).mean().round(1)
```

```python
# strip out max and min
temp_monthly_mean = temp_monthly_mean[[
    'Outside Temperature Baldy Summit (°F)',
    'Outside Temperature Baldy Draw (°F)',
    'Outside Temperature Sanfoin Bench (°F)',
    'Outside Temperature South Baldy (°F)',
    'Outside Temperature Orchard House (°F)',
    'Outside Temperature Indian Ridge (°F)',
 ]
]
```

```python
# start with mean for the month of daily average
output = 'output/orchard_house-temp_monthly_mean.csv'
temp_monthly_mean.to_csv(output)
```
