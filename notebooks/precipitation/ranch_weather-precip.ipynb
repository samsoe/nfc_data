```python
from statistics import mean
import pandas as pd
```

```python
full_src = '../src/orchard_house/orchard_house-export-full_history_001-190422.csv'
```

```python
full = pd.read_csv(full_src, parse_dates=['Date Time '], index_col=['Date Time '])
```

```python
full.info()
```

## Select for Precipitation

```python
# Select for Rain
precipi = full.loc[:,['Rain Amount TX1 (in)'
                    , 'Rain Amount TX2 (in)'
                    , 'Rain Amount TX3 (in)'
                    , 'Rain Amount TX4 (in)'
                    , 'Rain Amount TX5 (in)'
                    , 'Rain Amount TX6 (in)'
           ]].copy()
```

```python
precipi.info()
```

```python
# Rename columns for clarity
precipi.columns = ['Baldy Summit Rain (in)'
                 , 'Baldy Draw Rain (in)'
                 , 'Sanfoin Bench Rain (in)'
                 , 'South Baldy Ridge Rain (in)'
                 , 'Orchard House Rain (in)'
                 , 'Indian Ridge Rain (in)']
```

```python
precipi.tail()
```

```python
precipi['Baldy Summit Rain (in)'].sum()
```

```python
precipi = precipi.sort_index()
```

```python
precipi_month_sum = precipi.groupby(pd.Grouper(freq='M')).sum()
```

```python
precipi_year_sum = precipi.groupby(pd.Grouper(freq='Y')).sum()
```

```python
precipi_year_sum
```

```python
orchard_house_month = precipi['Orchard House Rain (in)'].groupby(pd.Grouper(freq='M')).sum()
```

```python
orchard_house_month.tail()
```

## Outputs

```python
precipi.head()
```

```python
output_src = '../output/orchard_house/orchard_house-database-190422.csv'
precipi.to_csv(output_src)
```

```python
# monthly sum aggregation
output_src = '../output/orchard_house/monthly_sum-database-190422.csv'
precipi_month_sum.to_csv(output_src)
```

```python
# yearly sum aggregation
output_src = '../output/orchard_house/yearly_precip_sum-database-190422.csv'
precipi_year_sum.to_csv(output_src)
```
